```python
import pandas as pd
import os

# Chemin vers le fichier
chemin_fichier = "/Users/ludovicveltz/Documents/Bootcamp_GENAI_2025/Crashcourse/hackaton_projet_W3D5/DATASET/plats_populaires_ingredients.csv"

def extraire_ingredients():
    # Lecture du fichier CSV
    print("Lecture du fichier...")
    df = pd.read_csv(chemin_fichier, sep=',')
    
    # Création d'un ensemble pour stocker les ingrédients uniques
    ingredients_uniques = set()
    
    # Parcours des ingrédients
    print("Extraction des ingrédients...")
    for index, row in df.iterrows():
        if pd.notna(row['Ingrédients']):
            # Split sur | et nettoyage de chaque ingrédient
            ingredients = [ing.split('(')[0].strip() for ing in row['Ingrédients'].split('|')]
            ingredients_uniques.update(ingredients)
    
    # Conversion en liste triée
    ingredients_liste = sorted(list(ingredients_uniques))
    
    print(f"\nNombre total d'ingrédients uniques trouvés : {len(ingredients_liste)}")
    print("\nListe des ingrédients :")
    for ing in ingredients_liste:
        print(f"- {ing}")
        
    return ingredients_liste

# Exécution
ingredients = extraire_ingredients()
```
